# Part I: Hypothesis testing

```r
n <- 5  #id-number

ak_vector <- c()

#generate ak values
for(k in 1:150) {
    ak <- k * log(k^2 * n + pi)
    x <- ak - floor(ak)
    ak_vector[k] <- x 
}

length(ak_vector)
# Generate X1,... X100 samples
x <- qnorm(ak_vector[1:100])
# Generate Y1,... X50 samples
y <- qnorm(ak_vector[101:150])
x
y
```

### Problem 1. H 0 : μ 1 = 0 vs. H 1 : μ < 0;σ1^ 2 is unknown.

Standard test:
The t test compares two averages (means) and tells if they are different from each other. The t test also tells how significant the differences are.
t-test T-test is used for the analysis of the two means. This test is good for this case since we have two independentsamples both sampled from a Normal distribution and we need to test their means when the variances are unknown.


```r
t_test = t.test(x, y, alternative = "less", mu = 0, paired = FALSE, var.equal = FALSE, conf.level = 0.95)
t_test
```

```r
print("Critical value: ")
alpha <- .05 
t_test.alpha = qt(1-alpha, df=149) 
t_test.alpha
```

```r
t_test.half.alpha = qt(1-alpha/2, df=149)
c(-t_test.half.alpha, t_test.half.alpha) 
```

The test statistic -0.0370 lies between the critical values -1.9760, and 1.97601. Hence, at .05 significance level, we do not reject the null hypothesis.

```r
t_test$p.value
```

If the P value is less than the chosen significance level -> reject the null hypothesis.
Significance level is 0.05, the corresponding confidence level is 95%.
P-value is 0.4853 and it's bigger than significance level, so H0 should be accepted!


### Problem 2. H 0 : μ 1 = μ 2 vs. H 1 : μ 1 != μ 2 ; σ 1 ^2 = σ 2^ 2 = 2

```r
#t.test(x, y, var.equal=TRUE)
ttest <- t.test(x, y, var.equal=TRUE, conf.level=0.95)
ttest
```

H0 : u1 = u2 (H0: means are equal)
H1: )alternative) means are not equal

Standard test: two-sided t-test. Can be used for the cases with two independent population samples both sampled from Normal distribution and when the null hypothesis is that the two means are equal, and the alternative is that they are not.

95 percent confidence interval: (-0.3372307, 0.3241764).

Significance level 0.05: t = -0.039003, test statistics falls in the rejection region(outside the confidence interval). The null hypothesis should be rejected.

p-value = 0.9689 larger than the significance level, thus we accept H0.
